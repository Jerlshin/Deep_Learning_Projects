```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Define the paths to your output images
style1_images = [
    "./Outputs/image_0.jpg",
    "./Style Image.jpg",
    "./Outputs/image_2500.jpg"
]

style2_images = [
    "./Stone_Outputs/image_0.jpg",
    "./stone_style.jpg",
    "./Stone_Outputs/image_2500.jpg"
]

# Create a 2x3 grid for plotting
fig, axes = plt.subplots(2, 3, figsize=(12, 8))

# Labels for the images
labels = ["Content Image", "Style Image", "Generated Image"]

# Loop through and display the images
for i, image_path in enumerate(style1_images):
    row = i // 3
    col = i % 3
    img = mpimg.imread(image_path)
    axes[row, col].imshow(img)
    axes[row, col].set_title(labels[i % 3])
    axes[row, col].axis('off')

for i, image_path in enumerate(style2_images):
    row = 1  # Second row
    col = i
    img = mpimg.imread(image_path)
    axes[row, col].imshow(img)
    axes[row, col].set_title(labels[i % 3])
    axes[row, col].axis('off')

# Adjust spacing and display the plot
plt.tight_layout()
plt.show()
```
